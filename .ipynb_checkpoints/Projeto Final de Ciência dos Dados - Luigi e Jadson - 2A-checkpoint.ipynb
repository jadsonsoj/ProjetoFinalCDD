# Treinando

```python
three_means = cluster.MiniBatchKMeans(n_clusters=4)
```

```python
three_means.fit(dadoscluster)
```

```python
saida = three_means.predict(dadoscluster)
```

```python
saida == 0
```

```python
conjunto = dadoscluster.copy()
```

```python
conjunto["saida"] = saida
```

crosstab - preferred e saida 

```python
pd.crosstab(conjunto['saida'],dados['Preferred Positions'])
```

```python
d0 = dados[saida == 0][['Acceleration', 'Dribbling']]
d1 = dados[saida == 1][['Acceleration', 'Dribbling']]
d2 = dados[saida == 2][['Acceleration', 'Dribbling']]

plt3d = plt.figure().gca(projection="3d")

plt3d.scatter(d0.Acceleration[:100], d0.Dribbling[:100], color="r")
plt3d.scatter(d1.Acceleration[:100], d1.Dribbling[:100], color="g")
plt3d.scatter(d2.Acceleration[:100], d2.Dribbling[:100], color="b")
```
